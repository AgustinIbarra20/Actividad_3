```python
#carga de un archivo csv sin √≠dices
df=pd.read_csv('Act2_1_Valores_nulos.csv')
df.head()
```

```python
#identificar valores nulos por columna
valores_nulos=df.isnull().sum()
valores_nulos
```

```python
#M√©todo aplicando desviaci√≥n est√°ndar. Encuentro los valores extremos en ventas_precios_constantes
y=df['ventas_precios_constantes']
Limite_superior= y.mean() + 3*y.std()
Limite_inferior= y.mean() - 3*y.std()
print('Limite superior permitido', Limite_superior)
print('Limite inferior permitido', Limite_inferior)
```

```python
#Encontramos outliers
outliers = df[(y>Limite_superior) | (y<Limite_inferior)]
outliers
```

```python
#obtenemos datos limpios
data_clean_ventas_precios_constantes = df[(y<=Limite_superior) & (y>=Limite_inferior)]
data_clean_ventas_precios_constantes
```

```python
#realizamos el diagrama de caja o bigote
fig = plt.figure(figsize =(10,6))
plt.boxplot(data_clean_ventas_precios_constantes['ventas_precios_constantes'])
plt.title('Outliers de ventas_precios_constantes')
plt.show() #dibujamos el diagrama
```

```python
#Metodo applicando cuartiles. Encuentro cuartiles 0.25 y 0.75

y=df['salon_ventas']

percentile25=y.quantile(0.25) #Q1
percentile75=y.quantile(0.75) #Q2
iqr= percentile75 - percentile25

Limite_Superior_iqr = percentile75 + 1.5*iqr
Limite_Inferior_iqr = percentile25 - 1.5*iqr
print('Limite superior permitido', Limite_Superior_iqr)
print('Limite inferior permitido', Limite_Inferior_iqr)
```

```python
#Encontramos outliers
outliers_iqr = df[(y>Limite_Superior_iqr) | (y<Limite_Inferior_iqr)]
outliers_iqr
```

```python
#obtenemos datos limpios
data_clean_salon_ventas = df[(y<=Limite_Superior_iqr) & (y>=Limite_Inferior_iqr)]
data_clean_salon_ventas
```

```python
#realizamos el diagrama de caja o bigote
fig = plt.figure(figsize =(10,6))
plt.boxplot(data_clean_salon_ventas['salon_ventas'])
plt.title('Outliers de salon_ventas')
plt.show() #dibujamos el diagrama
```
